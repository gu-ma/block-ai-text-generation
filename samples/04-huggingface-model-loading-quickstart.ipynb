**Big idea:** load one small model quickly and test generation behavior.

## What you will do

1. Load `distilgpt2`
2. Generate text from one prompt
3. Compare low vs high temperature
4. Relate outputs to earlier GPT4All notebooks

```python
# Install once if needed:
# !pip install -q transformers torch

# Import the text-generation pipeline from Hugging Face transformers
from transformers import pipeline
```

```python
# Load distilgpt2: a small, fast GPT2 model from Hugging Face
# First run will download the model (~500MB)
generator = pipeline("text-generation", model="distilgpt2")
print("distilgpt2 loaded")
```

## Step 1 — First generation

```python
# Generate text from a prompt
# do_sample=True enables diversity; temperature controls randomness
# temperature=0.8: moderate randomness (somewhat predictable but creative)
prompt = "In this class, AI can help students by"
out = generator(prompt, max_new_tokens=40, do_sample=True, temperature=0.8)
print(out[0]["generated_text"])
```

## Step 2 — Temperature comparison

```python
# Compare low vs high temperature on the same prompt
# Low temp (0.3): model favors most likely next tokens (conservative, repetitive)
# High temp (1.1): model spreads probability across more tokens (diverse, risky)
out_low = generator(prompt, max_new_tokens=40, do_sample=True, temperature=0.3)
out_high = generator(prompt, max_new_tokens=40, do_sample=True, temperature=1.1)

print("=== Temp 0.3 (more conservative) ===\n")
print(out_low[0]["generated_text"])
print("\n\n=== Temp 1.1 (more diverse) ===\n")
print(out_high[0]["generated_text"])
```
