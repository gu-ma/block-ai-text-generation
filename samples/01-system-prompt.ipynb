List all available models

```python
# List all models available locally
# Format: model_name, filename (GGUF format)
for model in GPT4All.list_models():
    print(f"{model['name']}, {model['filename']}")
```

```python
# Load the model from disk (quantized GGUF format)

# Change MODEL_FILENAME to match a model from the list above
MODEL_FILENAME = "Nous-Hermes-2-Mistral-7B-DPO.Q4_0.gguf"
model = GPT4All(MODEL_FILENAME)
print("Model loaded")
```

## Step 1 — Baseline (no system prompt)

We start with no role and no persona. This gives us a control output to compare later.

```python
# Define the user question we'll test with different system prompts
user_prompt = "Give me 5 project ideas for first-year design students using AI text tools."

# Generate baseline response with NO system prompt (empty string)
# This gives us a control to compare against
with model.chat_session(system_prompt=""):
    baseline = model.generate(user_prompt, max_tokens=180)
```

Helper to clean output

```python
# Helper function to clean up model output by removing empty lines
def clean_output(text):
    """Remove empty lines and trim whitespace from text"""
    output = []
    # Iterate through each line
    for line in text.splitlines():
        # Only keep non-empty lines
        if line:
            output.append(line.strip())
    # Rejoin into single string with clean line breaks
    return "\n".join(output)
```

```python
# Clean up baseline output
baseline = clean_output(baseline)

# Display baseline output for comparison
print("=== BASELINE ===")
print(baseline)
```

### What to observe

- Is the answer generic or specific?
- Does it follow a clear format?
- Is the tone suitable for your class context?

## Step 2 — Persona switcher (from list)

Try: `teacher`, `creative_director`, `critic`, `technical_assistant`.

```python
# Dictionary of different personas to try
# Each changes tone, structure, and focus
personas = {
    "teacher": "You are a patient teacher. Explain clearly with simple language.",
    "creative_director": "You are a creative director. Be inspiring, visual, and concise.",
    "critic": "You are a strict critic. Identify weak points and suggest improvements.",
    "technical_assistant": "You are a precise technical assistant. Structured and concise."
}

# Select which persona to use (try changing this value!)
selected_persona = "teacher"
system_prompt = personas[selected_persona]
print("Using persona:", selected_persona)
```

```python
# Generate response with the selected persona system prompt
with model.chat_session(system_prompt=system_prompt):
    persona_output = model.generate(user_prompt, max_tokens=180)

# Clean up persona output
persona_output = clean_output(persona_output)

# Display side-by-side comparison
print("=== BASELINE ===")
print(baseline)
print("\n=== PERSONA OUTPUT ===")
print(persona_output)
```

### What changed, exactly?

Compare baseline vs persona output on: tone, structure, specificity, and constraints.

If the factual content is similar but style changes a lot, that is expected: system prompts primarily steer behavior and framing.

## Step 3 — Prompt contract

A contract is a compact system instruction with:
- role
- audience
- constraints
- output format

```python
# Define a "prompt contract" - a structured system instruction
# Contracts control reliability and output format (not just tone)
prompt_contract = """
Role: course assistant
Audience: first-year art and design students
Constraints: simple language, max 120 words, no jargon
Output format: bullet list with 5 items + 1 caution note
""".strip()

# Generate response following the contract
with model.chat_session(system_prompt=f"Follow this contract exactly:\n{prompt_contract}"):
    contract_output = model.generate(user_prompt, max_tokens=180)

# Clean up contract output
contract_output = clean_output(contract_output)

# Display side-by-side comparison
print("=== BASELINE ===")
print(baseline)
print("\n=== PROMPT CONTRACT ===")
print(prompt_contract)
print("\n=== CONTRACT OUTPUT ===")
print(contract_output)
```

### Why contracts are useful

Personas control voice. Contracts control reliability and classroom usability.

Use personas for tone, and contracts for repeatable output quality.
